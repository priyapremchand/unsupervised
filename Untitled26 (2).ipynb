1. Loading and Preprocessing

```python
import pandas as pd
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]  
```

```python
print (iris_df.head())
```

```python
X = iris_df.drop('species', axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

2.Clustering Algorithm Implementation


A) KMeans Clustering 

KMeans clustering is an iterative algorithm that partitions a dataset into K clusters

It is a centroid-based algorithm, where each cluster is associated with a centroid.

The main aim of this algorithm is to minimize the sum of distances between the data point and their corresponding clusters.

Calculate the new centroids by averaging the data points in each cluste

choose k initals randomly

it will repeat assignment and update steps

suitability :-

find distinct clusters based on the feature measurements of the iris flowers.

The dataset is relatively small and well-separated, making it easier for KMeans to converge to an optimal solution


```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X)
```

```python
def plot_kmeans_clusters(X, labels):
    plt.figure(figsize=(8, 6))
    plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k')
    plt.title('KMeans Clustering of Iris Dataset')
    plt.xlabel(iris.feature_names[0])
    plt.ylabel(iris.feature_names[1])
    plt.grid(True)
    plt.show()

```

```python

plot_kmeans_clusters(X, kmeans_labels)
```

B) Hierarchical Clustering 

Hierarchical clustering creates a hierarchy of clusters by:

Agglomerative which Start with individual points and iteratively merge clusters based on a distance metric until one cluster remains.

Divisive which Start with one cluster and recursively split it into smaller clusters.

Suitability :

Hierarchical clustering is suitable becoz it allows for the visualization of the clustering process through a dendrogram.

It can reveal the relationships between clusters at different levels of granularity.

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
hierarchical = AgglomerativeClustering(n_clusters=3)
hierarchical_labels = hierarchical.fit_predict(X)
```

```python
def plot_hierarchical_clusters(X, labels):
    plt.figure(figsize=(8, 6))
    plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k')
    plt.title('Hierarchical Clustering of Iris Dataset')
    plt.xlabel(iris.feature_names[0])
    plt.ylabel(iris.feature_names[1])
    plt.grid(True)
    plt.show()
```

```python

plot_hierarchical_clusters(X, hierarchical_labels)
```

```python

plt.figure(figsize=(10, 7))
linked = linkage(X_scaled, method='ward')
dendrogram(linked, orientation='top', labels=range(len(X_scaled)))
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()
```
